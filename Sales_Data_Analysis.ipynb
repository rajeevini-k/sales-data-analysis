```python
# ==========================================
# SALES DATA ANALYSIS PROJECT
# Author: Rajeevini K
# Technologies: Python, Pandas, NumPy, Matplotlib
# ==========================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# ==========================================
# STEP 1: LOAD OR CREATE DATASET
# ==========================================

if os.path.exists("sales_data.csv"):
    df = pd.read_csv("sales_data.csv")
    print("Existing dataset loaded successfully.")
else:
    sales_data = {
        "Order_ID":[1001,1002,1003,1004,1005,1006,1007,1008,1009,1010,
                    1011,1012,1013,1014,1015,1016,1017,1018,1019,1020],

        "Date":[
            "2025-01-01","2025-01-02","2025-01-03","2025-01-05","2025-01-06",
            "2025-01-07","2025-01-10","2025-01-12","2025-01-15","2025-01-18",
            "2025-01-20","2025-01-22","2025-02-01","2025-02-03","2025-02-05",
            "2025-02-08","2025-02-10","2025-02-12","2025-02-15","2025-02-18"
        ],

        "Product":[
            "Laptop","Mobile","Headphones","Laptop","Mouse",
            "Keyboard","Mobile","Laptop","Headphones","Mobile",
            "Keyboard","Mouse","Laptop","Mobile","Headphones",
            "Keyboard","Mouse","Laptop","Mobile","Headphones"
        ],

        "Category":[
            "Electronics","Electronics","Accessories","Electronics","Accessories",
            "Accessories","Electronics","Electronics","Accessories","Electronics",
            "Accessories","Accessories","Electronics","Electronics","Accessories",
            "Accessories","Accessories","Electronics","Electronics","Accessories"
        ],

        "City":[
            "Hyderabad","Bangalore","Chennai","Hyderabad","Bangalore",
            "Chennai","Hyderabad","Mumbai","Mumbai","Delhi",
            "Delhi","Hyderabad","Chennai","Bangalore","Delhi",
            "Mumbai","Hyderabad","Delhi","Mumbai","Chennai"
        ],

        "Quantity":[
            2,3,5,1,8,
            4,2,3,7,4,
            6,10,2,5,6,
            4,9,2,3,5
        ],

        "Price":[
            55000,25000,2000,55000,800,
            1500,25000,55000,2000,25000,
            1500,800,55000,25000,2000,
            1500,800,55000,25000,2000
        ]
    }

    df = pd.DataFrame(sales_data)
    df.to_csv("sales_data.csv", index=False)
    print("Sample dataset created successfully.")

print("\nFirst 5 Records")
print(df.head())

# ==========================================
# STEP 2: DATA PREPARATION
# ==========================================

df["Date"] = pd.to_datetime(df["Date"])

df["Revenue"] = df["Quantity"] * df["Price"]

df["Month"] = df["Date"].dt.strftime("%B")

print("\nDataset Shape")
print(df.shape)

print("\nDataset Information")
print(df.info())

print("\nMissing Values")
print(df.isnull().sum())

# ==========================================
# STEP 3: DESCRIPTIVE STATISTICS
# ==========================================

print("\nSummary Statistics")
print(df.describe())

# ==========================================
# STEP 4: TOTAL REVENUE
# ==========================================

total_revenue = df["Revenue"].sum()

print("\n=================================")
print("TOTAL REVENUE")
print("=================================")
print(f"₹{total_revenue:,.2f}")

# ==========================================
# STEP 5: PRODUCT ANALYSIS
# ==========================================

product_sales = (
    df.groupby("Product")["Revenue"]
      .sum()
      .sort_values(ascending=False)
)

print("\n=================================")
print("REVENUE BY PRODUCT")
print("=================================")
print(product_sales)

# ==========================================
# STEP 6: CITY ANALYSIS
# ==========================================

city_sales = (
    df.groupby("City")["Revenue"]
      .sum()
      .sort_values(ascending=False)
)

print("\n=================================")
print("REVENUE BY CITY")
print("=================================")
print(city_sales)

# ==========================================
# STEP 7: CATEGORY ANALYSIS
# ==========================================

category_sales = (
    df.groupby("Category")["Revenue"]
      .sum()
      .sort_values(ascending=False)
)

print("\n=================================")
print("REVENUE BY CATEGORY")
print("=================================")
print(category_sales)

# ==========================================
# STEP 8: MONTHLY ANALYSIS
# ==========================================

monthly_sales = (
    df.groupby("Month")["Revenue"]
      .sum()
)

print("\n=================================")
print("MONTHLY REVENUE")
print("=================================")
print(monthly_sales)

# ==========================================
# STEP 9: VISUALIZATION
# ==========================================

plt.figure(figsize=(8,5))
monthly_sales.plot(kind="bar", color="skyblue")
plt.title("Monthly Revenue")
plt.ylabel("Revenue")
plt.xlabel("Month")
plt.tight_layout()
plt.savefig("monthly_sales.png")
plt.show()

plt.figure(figsize=(8,5))
product_sales.plot(kind="bar", color="green")
plt.title("Revenue by Product")
plt.ylabel("Revenue")
plt.xlabel("Product")
plt.tight_layout()
plt.savefig("top_products.png")
plt.show()

plt.figure(figsize=(8,5))
city_sales.plot(kind="bar", color="orange")
plt.title("Revenue by City")
plt.ylabel("Revenue")
plt.xlabel("City")
plt.tight_layout()
plt.savefig("sales_by_city.png")
plt.show()

plt.figure(figsize=(8,5))
category_sales.plot(
    kind="pie",
    autopct="%1.1f%%"
)
plt.title("Revenue Share by Category")
plt.ylabel("")
plt.tight_layout()
plt.savefig("category_share.png")
plt.show()

# ==========================================
# STEP 10: BUSINESS INSIGHTS
# ==========================================

top_product = product_sales.idxmax()
top_product_revenue = product_sales.max()

lowest_product = product_sales.idxmin()
lowest_product_revenue = product_sales.min()

top_city = city_sales.idxmax()
top_city_revenue = city_sales.max()

lowest_city = city_sales.idxmin()
lowest_city_revenue = city_sales.min()

top_category = category_sales.idxmax()
top_category_revenue = category_sales.max()

product_contribution = (
    top_product_revenue / total_revenue
) * 100

city_contribution = (
    top_city_revenue / total_revenue
) * 100

if len(monthly_sales) > 1:
    first_month_revenue = monthly_sales.iloc[0]
    last_month_revenue = monthly_sales.iloc[-1]

    growth_percentage = (
        (last_month_revenue - first_month_revenue)
        / first_month_revenue
    ) * 100
else:
    growth_percentage = 0

print("\n=================================")
print("BUSINESS INSIGHTS")
print("=================================")

print(f"Total Revenue Generated: ₹{total_revenue:,.2f}")

print(
    f"Top Performing Product: {top_product} "
    f"(Revenue: ₹{top_product_revenue:,.2f})"
)

print(
    f"Top Performing City: {top_city} "
    f"(Revenue: ₹{top_city_revenue:,.2f})"
)

print(
    f"Top Revenue Category: {top_category} "
    f"(Revenue: ₹{top_category_revenue:,.2f})"
)

print(
    f"Lowest Performing Product: {lowest_product} "
    f"(Revenue: ₹{lowest_product_revenue:,.2f})"
)

print(
    f"Lowest Revenue City: {lowest_city} "
    f"(Revenue: ₹{lowest_city_revenue:,.2f})"
)

print("\nKey Findings")

print(
    f"1. {top_category} generated the highest revenue "
    f"of ₹{top_category_revenue:,.2f}."
)

print(
    f"2. {top_product} contributed "
    f"{product_contribution:.2f}% of total revenue."
)

print(
    f"3. {top_city} contributed "
    f"{city_contribution:.2f}% of overall sales."
)

print(
    f"4. {lowest_product} generated the lowest revenue "
    f"among all products."
)

if growth_percentage > 0:
    print(
        f"5. Revenue increased by "
        f"{growth_percentage:.2f}% during the analysis period."
    )
elif growth_percentage < 0:
    print(
        f"5. Revenue decreased by "
        f"{abs(growth_percentage):.2f}% during the analysis period."
    )
else:
    print(
        "5. Revenue remained stable during the analysis period."
    )

# ==========================================
# STEP 11: EXPORT REPORTS
# ==========================================

product_sales.to_csv("product_sales_report.csv")

city_sales.to_csv("city_sales_report.csv")

category_sales.to_csv("category_sales_report.csv")

monthly_sales.to_csv("monthly_sales_report.csv")

print("\nReports Exported Successfully")

print("""
Generated Files:
1. sales_data.csv
2. product_sales_report.csv
3. city_sales_report.csv
4. category_sales_report.csv
5. monthly_sales_report.csv
6. monthly_sales.png
7. top_products.png
8. sales_by_city.png
9. category_share.png
""")

print("\nProject Completed Successfully")
```
